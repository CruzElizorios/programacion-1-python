# **¿Cómo graficamos en python?**
**Seaborn** es una librería para Python que *permite generar fácilmente elegantes gráficos*. **Seaborn** esta basada en **matplotlib** y proporciona una interfaz de alto nivel que es realmente sencilla de aprender.


```python
#Para usar esta librería, vamos a incluir estas líneas de código
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
#En el presente documento se muestran ejemplos como guías para graficar datos
```

```python
#Gráfico de barras
#Ejemplo 1:
#Supongamos que estamos trabajando con datos sobre precios de alquileres en una ciudad y que tenemos calculado el precio promedio de alquiler por barrio.
#Imaginemos que la ciudad en cuestión tiene 4 barrios y que tenemos guardados el precio promedio de cada barrio en un diccionario, donde
#la clave es el nombre del barrio y el valor, el promedio de los valores de los alquileres en dicho barrio
#Por ejemplo:
prom_precio_barrio = {"Barrio 1":150, "Barrio 2":400,"Barrio 3":700,"Barrio 4":270} #diccionario con las claves que son los datos del eje x y los valores los datos en el eje y

#Para realizar el gráfico de barras, debemos primero setear dimensión del gráfico, orientación de las etiquetas, tamaño de fuente, etc.
#En las siguientes líneas seteamos los valores del gráfico:
plt.figure(figsize = (10,3)) #para dar las dimensiones de la figura (ojo que debe ir arriba)
plt.xticks(rotation = 90, fontsize=15); #rota las etiquetas del eje x y tamaño de la fuente de los valores del eje x
plt.yticks(fontsize=15); #tamaño de la fuente de los valores del eje y
plt.xlabel('Barrios',fontsize=20) #indica qué representa el eje x
plt.ylabel('Precio medio',fontsize=20) #indica qué representa el eje y
plt.title('Precio medio por barrio en la ciudad X',fontsize=25) #titulo del gráfico y tamaño de la fuente
#esta última instrucción es la que construye el gráfico tomando en cuenta los valores seteados anteriomente
plt.bar(prom_precio_barrio.keys(), prom_precio_barrio.values()) #pasamos los datos a graficar, en el eje x las claves del diccionario, en el y los valores del diccionario
```

```python
#Gráfico de barras
#Ejemplo 2:
#Podemos tener los datos en dos listas apareadas
lista_x =["Barrio 1","Barrio 2","Barrio 3","Barrio 4"]
lista_y =[150,400,700,270]


plt.figure(figsize = (15,6))
plt.title('Precio medio por barrio en X ciudad',fontsize=25) #titulo del gráfico y tamaño de la fuente
plt.bar(lista_x, lista_y)
#podemos modificar el gráfico a nuestro gusto como en el ejemplo anterior
```

```python
#Gráfico de barras
#Ejemplo 3:
#ahora modularizamos, seteamos dentro de una función las características que queremos para todo gráfico de
#barra que hagamos y dejamos como parámetros aquellos valores que van a cambiar

def dibujar_grafico_barra(titulo, etiqueta_x, etiqueta_y,datos_x,datos_y):
  plt.figure(figsize = (15,6)) #para dar las dimensiones de la figura (ojo que debe ir arriba)
  plt.xticks(rotation = 90,fontsize=15); #rota las etiquetas del eje x y tamaño de la fuente de los valores del eje x
  plt.yticks(fontsize=15); #tamaño de la fuente de los valores del eje y
  plt.xlabel(etiqueta_x,fontsize=20) #indica qué representa el eje x
  plt.ylabel(etiqueta_y,fontsize=20) #indica qué representa el eje y
  plt.title(titulo,fontsize=25) #titulo del gráfico y tamaño de la fuente
  plt.bar(datos_x, datos_y) #pasamos los datos a graficar

```

```python
dibujar_grafico_barra("Gráfico 1","Barrios", "Precio Promedio",lista_x,lista_y)
dibujar_grafico_barra("Gráfico 2","Barrios", "Precio Promedio",prom_precio_barrio.keys(),prom_precio_barrio.values())
```

```python
#Gráfico de torta para representar porcentajes
#Ejemplo:
#Siguiendo el mismo ejemplo de los alquileres, ahora vamos a considerar qué porcentaje de las ofertas de alquileres pertenecen a cada barrio.
#Guardamos la información en dos listas apareadas
lista_barrios = ["Barrio 1","Barrio 2","Barrio 3","Barrio 4"]
lista_porcent =[0.35,0.40,0.15,0.1] #los porcentajes deben sumar 1

def dibujar_grafico_torta(titulo, etiquetas,porcentajes):
  plt.figure(figsize = (15,6))
  plt.title(titulo,fontsize=25) #titulo del gráfico y tamaño de la fuente
  plt.pie(porcentajes, labels  = etiquetas, autopct="%0.1f %%") #autopct="%0.1f %%" hace que aparezcan los porcentajes dentro de cada parte de la torta
  #podemos setear todo lo que querramos como antes, tamaños de fuentes, título, etc.

dibujar_grafico_torta("Porcentajes",lista_barrios, lista_porcent)
```

```python
#Ahora vamos a ver cómo construir un Histograma

#Ejemplo 4:
#Supongamos que recolectamos datos sobre el consumo de dulces de un grupo de niños durante una semana y calculamos el promedio de
#consumo semanal de cada uno de ellos
consumo_dulces = [2,4,6,34,23,12,34,56,32,78,33,22,3,3,4,5,6,6,15,30,78] #cantidad de dulces promedio que consume cada niño entrevistado por semana

def dibujar_histograma(titulo,etiqueta_x,etiqueta_y, datos):
    plt.figure(figsize = (15,6)) #para dar las dimensiones de la figura (ojo que debe ir arriba)
    plt.title(titulo ,fontsize=25)
    plt.xlabel(etiqueta_x,fontsize=20) #indica qué representa el eje x
    plt.ylabel(etiqueta_y,fontsize=20) #indica qué representa el eje y
    plt.hist(datos, bins = 5) #la cantidad de bins refiere a la cantidad de rangos en los cuales se van a dividir los datos para graficar
    #podemos setear todo lo que querramos como antes, tamaños de fuentes, etc.

dibujar_histograma("Consumo promedio de dulces por semana","Rangos de consumo", "Cantidad de niños",consumo_dulces)
```

```python
#Graficar puntos geoespaciales usando scatter plot

#Ejemplo 5:

#Tomamos 20 departamentos de una ciudad cualquiera para graficar

#latitudes de los puntos a graficar
ejemplo_lats = [52.36435, 52.36407, 52.38761, 52.36775, 52.36584, 52.37168, 52.36786,52.38028, 52.36158, 52.36811,
                52.358112, 52.35597, 52.36431, 52.37828, 52.36968, 52.35774, 52.34801, 52.35665, 52.35178]
#longitudes de los puntos a graficar
ejemplo_lons = [4.94358, 4.89393, 4.89188, 4.89092, 4.89111, 4.91471, 4.87458, 4.89089, 4.86101, 4.88262, 4.897637,
                4.88512, 4.93122, 4.8944, 4.88122, 4.89647, 4.83639, 4.86363, 4.90699]

def dibujar_puntos(titulo,longitudes, latitudes):
  plt.figure(figsize = (15,8))
  plt.xlabel('Longitud',fontsize=20)
  plt.ylabel('Latitude',fontsize=20)
  plt.title(titulo,fontsize=25)
  plt.scatter(longitudes, latitudes)


dibujar_puntos("Ubicación ofertas de alquileres",ejemplo_lons, ejemplo_lats)
```

```python
# Crear gráfico de líneas

# Ejemplo 6 - Versión 1:
# Supongamos que disponemos de datos de ventas mensuales correspondientes a un año, de una empresa
# Podemos construir un gráfico de lineas que los modele

# Supongamos que los datos disponibles corresponden al año 2000
anio = 2000

# Disponemos los datos que queremos graficar en un diccionario (los meses en que no hubo ventas no figuran en el diccionario)
datos_anio = {"Febrero": 400,  "Abril": 300, "Marzo": 600, "Enero": 200,  "Junio": 555, "Octubre": 300, "Noviembre": 800}

plt.figure(figsize=(10, 5))
plt.plot(datos_anio.keys(), datos_anio.values(), marker='o', label=f'Año {anio}')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title(f'Ventas mensuales correspondientes al año {anio}')
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# En el gráfico anterior los meses se graficaron en el orden que aparecían en el diccionario... resulta conveniente?
# Podríamos ordenarlos?
# Probemos ordenar los meses alfabéticamente

# Ejemplo 6 - Versión 2:

# Supongamos que los datos disponibles corresponden al año 2000
anio = 2000

# Disponemos los datos que queremos graficar en un diccionario (los meses en que no hubo ventas no figuran en el diccionario)
datos_anio = {"Febrero": 400,  "Abril": 300, "Marzo": 600, "Enero": 200,  "Junio": 555, "Octubre": 300, "Noviembre": 800}

# la diferencia con la versión anterior está aquí
# Creamos una lista con todas las claves del diccionario
lista_meses = list(datos_anio.keys())
lista_meses.sort()

# Creamos una lista con valores para graficar teniendo en cuenta el orden de los meses
ventas_anio = []
for mes in lista_meses:
  ventas_anio.append(datos_anio.get(mes))

plt.figure(figsize=(10, 5))
plt.plot(lista_meses, ventas_anio, marker='o', label=f'Año {anio}')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title(f'Ventas mensuales correspondientes al año {anio}')
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Los meses ahora se grafican en el orden alfabético ... resulta conveniente?
# Claro que no!
# Necesitaríamos ver la evolución temporal e incluir incluso los meses en los cuales no hubo ventas
# Mostramos una nueva solución al problema

# Ejemplo 6 - Versión 3:

# Supongamos que los datos disponibles corresponden al año 2000
anio = 2000

# Disponemos los datos que queremos graficar en un diccionario (los meses en que no hubo ventas no figuran en el diccionario)
datos_anio = {"Febrero": 400,  "Abril": 300, "Marzo": 600, "Enero": 200,  "Junio": 555, "Octubre": 300, "Noviembre": 800}

# Creamos una lista con todos los meses del año
lista_meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre",
               "Octubre", "Noviembre", "Diciembre"] # la diferencia con la versión anterior está aquí
                                                    # Usamos esta lista predeterminada

# Creamos una lista con valores para graficar teniendo en cuenta el orden de los meses
ventas_anio = []
# Recordemos el comportamiento get() de diccionarios:
# - Si la clave que buscamos existe, nos devuelve su valor.
# - Si la clave que buscamos no existe, nos devuelve None salvo que le indiquemos otro valor por defecto
for mes in lista_meses:
  ventas_anio.append(datos_anio.get(mes,0)) #si el mes no existe en el diccionario, a ese mes le asignamos ventas 0

plt.figure(figsize=(10, 5))
plt.plot(lista_meses, ventas_anio, marker='o', label=f'Año {anio}')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title(f'Ventas mensuales correspondientes al año {anio}')
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Cuando graficamos vemos que los meses se grafican en orden cronológico y que los meses en 0 también aparecen en el gráfico
# Ahora, qué pasa si disponemos datos de dos años y queremos compararlos?

# Ejemplo 6 - Versión 4:

# Supongamos que los datos disponibles corresponden a los años 2000 y 2001
anio1 = 2000
anio2 = 2001

# Disponemos los datos que queremos graficar en diccionarios (los meses en que no hubo ventas no figuran en los diccionarios)
datos_anio1 = {"Enero": 200, "Febrero": 400, "Marzo": 600, "Abril": 300, "Junio": 555, "Octubre": 300, "Noviembre": 800}
datos_anio2 = {"Enero": 100, "Febrero": 300, "Julio": 500, "Septiembre": 200, "Octubre": 600, "Diciembre":900 }

# Creamos una lista con todos los meses del año
lista_meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre",
               "Octubre", "Noviembre", "Diciembre"]

# Creamos listas con valores para graficar teniendo en cuenta el orden de los meses
ventas_anio1 = []
ventas_anio2 = []
for mes in lista_meses:
  ventas_anio1.append(datos_anio1.get(mes,0)) #si el mes no existe en el diccionario, a ese mes le asignamos ventas 0
  ventas_anio2.append(datos_anio2.get(mes,0)) #si el mes no existe en el diccionario, a ese mes le asignamos ventas 0


plt.figure(figsize=(10, 5))
plt.plot(lista_meses, ventas_anio1, marker='o', label=f'Año {anio1}')
plt.plot(lista_meses, ventas_anio2, marker='o', label=f'Año {anio2}')
plt.xlabel('Meses')
plt.ylabel('Ventas')
plt.title(f'Comparación de ventas por mes: {anio1} vs {anio2}')
plt.xticks(rotation=60)
plt.legend()
plt.tight_layout()
plt.show()

# TENER EN CUENTA QUE EL CÓDIGO AQUÍ PRESENTADO NO ESTÁ MODULARIZADO
# Para usarlo deberían definir funciones
```
