```python
def ingresoDeDatos() -> list[(str,float)]:
  """
  Esta funcin solicita al usuario que ingrese los datos presentados
  en la planilla y devuelva una lista de tuplas, donde la primer
  componente es el nombre del barrio y la segunda componente es el
  precio por metro cuadrado del barrio
  """
  barrios = []

  while True:
    datoBarrio = input("Ingrese el nombre de un barrio de Rosario o salir para terminar: ")
    if(datoBarrio == "salir"):
      break
    datoPrecio = float(input(f"Ingrese el precio en dolares del barrio {datoBarrio}: "))
    tuplaDato = (datoBarrio,datoPrecio)
    barrios.append(tuplaDato)

  return barrios


def promMinMax(listaB: list[(str,float)]) -> tuple[float,str,str]:
  """
  Esta funcin recibe una lista de tuplas donde la primera componente
  es el nombre del barrio y la segunda componente es el precio por
  metro cuadrado del barrio, y devuelve el precio promedio en la
  ciudad , el nombre del barrio ms barato y el nombre del barrio
  ms caro
  """
  barrioMasBarato = listaB[0][0]
  barrioMasCaro = listaB[0][0]
  precioMin = listaB[0][1]
  precioMax = listaB[0][1]

  suma = 0

  for barrio, precio in listaB:
    suma+=precio
    if(precio < precioMin):
      barrioMasBarato = barrio
      precioMin = precio
    elif(precio > precioMax):
      barrioMasCaro = barrio
      precioMax = precio

  promedio = round(suma/len(listaB),2)
  # toma esto-> [('arroyito', 1200.0), ('las delicias', 3600.0), ('peras', 700.0), ('saladillo', 7500.0), ('san martin', 4620.5)]
  # devuelve -> (3524.1, 'peras', 'saladillo')
  return promedio , barrioMasBarato , barrioMasCaro

def principal() -> None:
  """
  Este es el programa principal
  """
  listaBarrios = ingresoDeDatos ()
  promedio, barrioBarato, barrioCaro = promMinMax(listaBarrios)
  print(f"""
El promedio del precio por metro cuadrado en la ciudad de Rosario es ${promedio}USD, el barrio ms caro de la ciudad es: {barrioCaro.capitalize()} y el ms barato es: {barrioBarato.capitalize()}
  """)

principal()
```
